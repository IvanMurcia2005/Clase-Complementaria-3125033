```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos de adopción de IA en diferentes sectores
data = {
    'Año': [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    'Comercio Electrónico (%)': [10, 15, 20, 30, 40, 50, 60, 70, 80],
    'Atención al Cliente (%)': [5, 8, 12, 20, 28, 35, 45, 55, 65],
    'Movilidad (Vehículos Autónomos) (%)': [1, 2, 3, 5, 8, 10, 12, 15, 18],
    'Salud (%)': [3, 5, 8, 12, 18, 25, 35, 40, 50],
    'Educación (%)': [1, 2, 3, 5, 8, 12, 18, 25, 35]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Crear el gráfico
plt.figure(figsize=(12, 8))

# Graficar cada sector
for column in df.columns[1:]:
    plt.plot(df['Año'], df[column], marker='o', label=column)

# Personalizar el gráfico
plt.title('Adopción de IA en Diferentes Sectores (2015-2023)', fontsize=15)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Adopción (%)', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
