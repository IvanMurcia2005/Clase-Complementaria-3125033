```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV
file_path = '/content/popularidad_lenguajes_programacion.csv'  # Ajusta la ruta si es necesario
df = pd.read_csv(file_path)

# Crear el gráfico de líneas
plt.figure(figsize=(12, 8))

# Graficar cada lenguaje
for column in df.columns[1:]:  # Excluye la columna 'Año'
    plt.plot(df['Año'], df[column], marker='o', label=column)

# Personalizar el gráfico
plt.title('Evolución de la Popularidad de Lenguajes de Programación (2010-2023)', fontsize=15)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Popularidad (%)', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
