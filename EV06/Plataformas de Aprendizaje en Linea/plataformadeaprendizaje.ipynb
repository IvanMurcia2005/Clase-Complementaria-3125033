```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos de usuarios activos en plataformas de aprendizaje en línea
data = {
    'Año': [2017, 2018, 2019, 2020, 2021, 2022, 2023],
    'Coursera (Millones)': [30, 36, 45, 65, 77, 90, 110],
    'Udemy (Millones)': [20, 24, 30, 50, 60, 75, 90],
    'edX (Millones)': [14, 16, 18, 24, 30, 36, 42],
    'Khan Academy (Millones)': [15, 18, 22, 30, 35, 40, 50],
    'LinkedIn Learning (Millones)': [6, 8, 10, 20, 25, 30, 40]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Crear el gráfico
plt.figure(figsize=(12, 8))

# Graficar cada plataforma
for column in df.columns[1:]:
    plt.plot(df['Año'], df[column], marker='o', label=column)

# Personalizar el gráfico
plt.title('Crecimiento de Plataformas de Aprendizaje en Línea (2017-2023)', fontsize=15)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Usuarios Activos (Millones)', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
