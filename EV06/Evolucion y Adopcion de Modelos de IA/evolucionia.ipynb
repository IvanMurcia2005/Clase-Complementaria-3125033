```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import io

# Datos
data = '''Año,Modelo de IA,Porcentaje de Adopcion
1956,Dartmouth Conference,0
1966,ELIZA,5
1997,Deep Blue,10
2011,Watson,15
2016,AlphaGo,20
2018,BERT,25
2020,GPT-3,35
2022,DALL-E 2,35
2023,ChatGPT,65
2024,"Claude, GPT-4",72'''

# Crear DataFrame
df = pd.read_csv(io.StringIO(data))

# Variables independientes y dependientes
X = df[['Año']]  # Variable independiente (Años)
y = df['Porcentaje de Adopcion']  # Variable dependiente

# Crear modelo de regresión lineal
model = LinearRegression()
model.fit(X, y)  # Ajustar el modelo

# Predicciones
y_pred = model.predict(X)

# Visualizar resultados
plt.figure(figsize=(12, 6))
plt.scatter(df['Año'], df['Porcentaje de Adopcion'], color='blue', label='Datos Reales')
plt.plot(df['Año'], y_pred, color='red', label='Regresión Lineal')
plt.title('Evolución y Adopción de Modelos de IA (Regresión Lineal)', fontsize=15)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Porcentaje de Adopción (%)', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Métricas de evaluación
print("Pendiente (Coeficiente):", model.coef_[0])
print("Intercepto:", model.intercept_)
print("Error Cuadrático Medio (MSE):", mean_squared_error(y, y_pred))
print("R² (Coeficiente de determinación):", r2_score(y, y_pred))
```
