```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos de ciberseguridad y ciberataques
data = {
    'Año': [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    'Inversión en Ciberseguridad (USD Billones)': [75, 87, 95, 106, 124, 142, 155, 170, 187],
    'Ciberataques Globales (Millones)': [8, 10, 13, 16, 20, 24, 29, 35, 40]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Crear el gráfico de barras para Inversión y Línea para Ciberataques
fig, ax1 = plt.subplots(figsize=(12, 8))

# Gráfico de barras para inversiones
ax1.bar(df['Año'], df['Inversión en Ciberseguridad (USD Billones)'], color='blue', alpha=0.7, label='Inversión en Ciberseguridad')
ax1.set_xlabel('Año', fontsize=12)
ax1.set_ylabel('Inversión en Ciberseguridad (USD Billones)', fontsize=12, color='blue')

# Crear un segundo eje Y para ciberataques
ax2 = ax1.twinx()
ax2.plot(df['Año'], df['Ciberataques Globales (Millones)'], color='red', marker='o', label='Ciberataques Globales')
ax2.set_ylabel('Ciberataques Globales (Millones)', fontsize=12, color='red')

# Añadir títulos y leyendas
plt.title('Inversión en Ciberseguridad y Ciberataques Globales (2015-2023)', fontsize=15)
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Mostrar el gráfico
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
