```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos de adopción de vehículos eléctricos
data = {
    'Año': [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    'Ventas Globales de VE (Millones)': [0.5, 0.8, 1.2, 1.8, 2.2, 3.1, 6.6, 10.2, 13.0],
    'Porcentaje de Crecimiento Anual (%)': [None, 60, 50, 50, 22, 41, 113, 55, 27]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Crear el gráfico de barras para Ventas y Línea para Crecimiento Anual
fig, ax1 = plt.subplots(figsize=(12, 8))

# Gráfico de barras para ventas globales
ax1.bar(df['Año'], df['Ventas Globales de VE (Millones)'], color='green', alpha=0.7, label='Ventas Globales de VE (Millones)')
ax1.set_xlabel('Año', fontsize=12)
ax1.set_ylabel('Ventas Globales de VE (Millones)', fontsize=12, color='green')

# Crear un segundo eje Y para el porcentaje de crecimiento anual
ax2 = ax1.twinx()
ax2.plot(df['Año'], df['Porcentaje de Crecimiento Anual (%)'], color='blue', marker='o', label='Crecimiento Anual (%)')
ax2.set_ylabel('Porcentaje de Crecimiento Anual (%)', fontsize=12, color='blue')

# Añadir títulos y leyendas
plt.title('Adopción Global de Vehículos Eléctricos (2015-2023)', fontsize=15)
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Mostrar el gráfico
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
