```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Ruta del archivo CSV (ajusta según el archivo que subas)
file_path = '/content/metaverso.csv'

# Leer el archivo CSV
df = pd.read_csv(file_path)

# Crear el gráfico
plt.figure(figsize=(14, 8))  # Tamaño del gráfico

# Graficar Usuarios Activos y Tamaño del Mercado
plt.plot(df['Año'], df['Usuarios Activos (Millones)'], marker='o', label='Usuarios Activos (Millones)', color='blue')
plt.plot(df['Año'], df['Tamaño del Mercado (USD Billones)'], marker='s', label='Tamaño del Mercado (USD Billones)', color='orange')

# Ajustar ticks del eje X para mostrar años
plt.xticks(df['Año'], fontsize=12)
plt.xlabel("Año", fontsize=14)

# Configurar ticks del eje Y con decimales redondeados
plt.yticks(fontsize=12)
plt.ylabel("Valor", fontsize=14)

# Títulos y etiquetas
plt.title("Evolución del Metaverso: Usuarios y Tamaño de Mercado", fontsize=18)

# Agregar leyenda
plt.legend(loc='upper left', fontsize=12)

# Cuadrícula para mayor claridad
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ajustar diseño y mostrar
plt.tight_layout()
plt.show()
```
