```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos de popularidad de librerías de programación
data = {
    'Año': [2018, 2019, 2020, 2021, 2022, 2023],
    'NumPy (%)': [25, 30, 40, 50, 60, 70],
    'Pandas (%)': [20, 25, 35, 45, 55, 65],
    'TensorFlow (%)': [30, 35, 45, 50, 60, 65],
    'PyTorch (%)': [15, 20, 25, 35, 45, 55],
    'React (%)': [40, 42, 45, 50, 55, 60],
    'Angular (%)': [35, 34, 33, 32, 30, 28]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Crear el gráfico
plt.figure(figsize=(12, 8))

# Graficar cada librería
for column in df.columns[1:]:
    plt.plot(df['Año'], df[column], marker='o', label=column)

# Personalizar el gráfico
plt.title('Popularidad de Librerías de Programación (2018-2023)', fontsize=15)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Popularidad (%)', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
